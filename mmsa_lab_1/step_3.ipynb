You need to make an estimation of order statistics and represent them as “box with
whiskers” plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

res = pd.read_csv('data/PS4_GamesSales.csv', header=0, encoding='unicode_escape')
# fixing some data
res.fillna(value='N\A', inplace=True)

# Creating dataset
data = [res['North America'], res['Europe'], res['Japan'], res['Rest of World']]

fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

# Creating axes instance
bp = ax.boxplot(data, patch_artist = True,
                notch ='True', vert = 0, whis=140.0)

colors = ['#0000FF', '#00FF00',
          '#FFFF00', '#FF00FF']

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)

# x-axis labels
ax.set_yticklabels(['North America', 'Europe',
                    'Japan', 'Rest of the world'])

# Adding title
plt.title("Sales by region")

# Removing top axes and right axes
# ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

# show plot
plt.show()
```
