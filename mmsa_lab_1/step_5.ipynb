# Step 5
    Estimate parameters of chosen distributions using methods of maximum likelihood
and least squares method.

```python
import scipy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

res = pd.read_csv('data/PS4_GamesSales.csv', header=0, encoding='unicode_escape')
# fixing some data
res.fillna(value='N\A', inplace=True)

na_mean = res['North America'].mean()
na_std = res['North America'].std()

# maximum likelihood method
# mu = mean value
mu_mle = na_mean
# sigma = standard deviation
sigma_mle = na_std

fit = scipy.stats.lognorm.fit(res['North America'])
# fit returns alpha, loc, beta
# parameters for North America
mu = fit[0]
sigma = fit[2]

print(mu_mle)
print(sigma_mle)
print(mu)
print(sigma)

# TODO probably need to google how to estimate lognorm distribution using those methods

# TODO probably it is lognormal distribution and following approach should be used:
# The distribution parameters are determined using the fit function based on the maximum likelihood method
# params = scipy.stats.lognorm.fit(df_curr_first['amount'])
# pdf = scipy.stats.lognorm.pdf(x, *params)

res['North America'].hist(density=True, bins=10, grid=False)
x = np.linspace(0, res['North America'].max())
# plt.plot(x, scipy.stats.lognorm.pdf(x, alpha_mom, beta_mom), 'm-')
plt.plot(x, scipy.stats.lognorm.pdf(x, mu, sigma), 'r--')
plt.plot(x, scipy.stats.lognorm.pdf(x, mu_mle, sigma_mle), 'g--')
```
